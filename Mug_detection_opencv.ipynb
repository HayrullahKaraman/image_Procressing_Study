```python
import numpy as np
import copy
import cv2
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
from PIL import Image
from keras import models
from tensorflow.keras.utils import img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix
import os
```

```python
model = models.load_model('mug2c.h5')
class_names  = {0: "mug", 1: "nomug"}
video = cv2.VideoCapture(0)
```

```python
while True:
        _, frame = video.read()

        #Convert the captured frame into RGB
        im = Image.fromarray(frame, 'RGB')

        #Resizing into dimensions you used while training
        im = im.resize((224,224))
       
        
        img_array = np.array(im)
       # x,y,w = img_array.shape
        #roi_gray = gray[y:y + h, x:x + w]
        #roi_color =img[y:y + h, x:x + w]
        
        #Expand dimensions to match the 4D Tensor shape.
        img_array = np.expand_dims(img_array, axis=0)
    
        #Calling the predict function using keras
        prediction = model.predict(img_array)#[0][0]
        mug = tf.nn.sigmoid(prediction[0])
        names=class_names[np.argmax(mug)]
        print(prediction)
        
        #Customize this part to your liking...
        
                     
        if(prediction[0][1]>0.5):
            print(names)
        elif(prediction[0][0]>0.5):
            print(names)

        cv2.imshow("Prediction", frame)
        key=cv2.waitKey(1)
        if key == 27:
            print('esc is pressed closing all windows')
            cv2.destroyAllWindows()
            break
        
    
               
if cv2.getWindowProperty('sunset', cv2.WND_PROP_VISIBLE) < 1:
    print("ALL WINDOWS ARE CLOSED")
cv2.waitKey(0)
cv2.destroyAllWindows()       
```
