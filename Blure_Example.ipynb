İmport Libary

```python
import numpy  as np
from matplotlib import pyplot as plt
import cv2
```

İmage ekleme

```python
img=cv2.imread('D:\hayrullah.jpeg',1)
```

Bluer Kernel uygulaması

```python
kernel=np.ones((19,19),np.float32)/225
hdf=cv2.filter2D(img,-1,kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(hdf),plt.title('Blur')
plt.xticks([]), plt.yticks([])
plt.show()
```
